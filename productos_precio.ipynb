```python
import pandas as pd

# Configuración para mostrar todos los resultados completos
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# Cargar el archivo CSV
df = pd.read_csv('products.csv')

# Reemplazar valores nulos en la columna 'Precio de venta' con 0 para simplificar las comparaciones
df['Precio de venta'].fillna(0, inplace=True)

# Identificar productos con precio igual a 0 (incluyendo los originalmente nulos)
productos_sin_precio = df[df['Precio de venta'] == 0]

print("\nProductos sin precio (0 o nulos):")
print(productos_sin_precio)

# Guardar los resultados en archivos CSV
df.to_csv('productos_sin_precio.csv', index=False)
```
